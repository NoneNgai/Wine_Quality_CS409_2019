```python
dataset = list([])
with open(r'C:\Users\Kittichai\test409\cs409\project\winequality-red.csv' ,mode='r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    line_count = 0
    for row in csv_reader:
        if line_count != 0: 
            dataset.append(row)
        line_count += 1

dataset = np.array(dataset).astype(np.float)

correlationR = np.ones([12,12])

for i in range(12):
    for j in range(12):
        correlationR[i][j] = np.sum((dataset[:,i] - np.mean(dataset[:,i])) *(dataset[:,j] - np.mean(dataset[:,j])) * (1 / (len(dataset) -1))) / (np.std(dataset[:,i]) * np.std(dataset[:,j]))

dataset = list([])
with open(r'C:\Users\Kittichai\test409\cs409\project\winequality-white.csv' ,mode='r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    line_count = 0
    for row in csv_reader:
        if line_count != 0: 
            dataset.append(row)
        line_count += 1

dataset = np.array(dataset).astype(np.float)

correlationW = np.ones([12,12])

for i in range(12):
    for j in range(12):
        correlationW[i][j] = np.sum((dataset[:,i] - np.mean(dataset[:,i])) *(dataset[:,j] - np.mean(dataset[:,j])) * (1 / (len(dataset) -1))) / (np.std(dataset[:,i]) * np.std(dataset[:,j]))

```

```python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
name = ["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","quality"]

my,mx=np.mgrid[slice(0,12,1),slice(0,12,1)]
fig= plt.figure(figsize=(10,8))
plt.pcolor(mx,my,correlationR,cmap=cm.seismic)
plt.colorbar()
plt.scatter(name, name ,color="maroon")
plt.suptitle('red wine qurality')
print(correlationR[11])
```

```python
fig= plt.figure(figsize=(10,8))
plt.pcolor(mx,my,correlationW,cmap=cm.BrBG)
plt.colorbar()
plt.scatter(name, name ,color="g")
plt.suptitle('white wine qurality')
print(correlationW[11])
```
